```python
import pandas as pd 
df=pd.read_excel(r"C:\Users\user\Desktop\Ironman Ratings.xlsx")
df
```

```python
Stars_list=df['No. of stars'].tolist()                        # CONVERTING THR STARS COLUMN TO LIST
print(Stars_list)
Voters_list=df['No. of Voters'].tolist()                # CONVERTING THR VOTERS COLUMN TO LIST
print(Voters_list)
sum_ratings=sum(Stars_list[i]*Voters_list[i] for i in range(len(Stars_list)))       # TOTAL 
print(sum_ratings)
```

```python
Total_voters=(df['No. of Voters']).sum()             # FREQUENCY OF VOTERS
print(Total_voters)
```

```python
mean=sum_ratings/Total_voters                      # CALCULATING MEAN
print("The Arithmetic Mean of the Movie Ironman is =",mean)
```

```python
data=list(Stars_list[i]*Voters_list[i] for i in range(len(Stars_list)))        # STORING IN LIST TO ZIP THE RECORD
print(data)
```

```python
data1=list(zip(Stars_list,Voters_list))
print(data1)
```

```python
data1.sort()
elements=len(data1)
print(elements)

if elements % 2 == 0:
    mid1=data1[elements//2-1]
    mid2=data1[elements//2]
    median = (mid1[0]+mid2[0])/2                  # [0] REPRESENTS THE INDEX 
else:
    median = data1[elements//2][0]
print("The Median of the Movie Ironman is =",median)
```

```python
#   Median is the midpoint. A quick way to get this is to average the highest and lowest number. 
#   So, say that you have those 100 reviews above, and the top review was 10 stars, 
#   and the bottom review was 1, then the median would be 5.5.

# So, we should never prefer median method over mean method while calculating the rating of a "MOVIE".
```

```python
# ORIGINAL MOVIE(Kaithi) 

import pandas as pd 
df1=pd.read_excel(r"C:\Users\user\Desktop\Kaithi ratings.xlsx")
df1
```

```python
Stars_list2=df1['Stars'].tolist()
print(Stars_list2)
Voters_list2=df1['No. of voters'].tolist()
print(Voters_list2)

sum_ratings2=sum(Stars_list2[i]*Voters_list2[i] for i in range(len(Stars_list2)))
print(sum_ratings2)
```

```python
Total_voters1=(df1['No. of voters']).sum()
print(Total_voters1)
```

```python
mean2=sum_ratings2/Total_voters1
print("The Arithmetic Mean of the Movie Kaithii is =",mean2)
```

```python
data_Ka=list(Stars_list2[i]*Voters_list2[i] for i in range(len(Stars_list2)))
print(data_Ka)
data_Kaithi=list(zip(Stars_list2,Voters_list2))
print(data_Kaithi)
```

```python
def mean(data_Kaithi):
    return sum([x[1] for x in data_Kaithi]) / len(data_Kaithi)

def variance(data_Kaithi):
    m = mean(data_Kaithi)
    return sum([(x[1] - m) ** 2 for x in data_Kaithi]) / len(data_Kaithi)

def std_deviation(data_Kaithi):
    return variance(data_Kaithi) ** 0.5

print("Standard deviation:", std_deviation(data_Kaithi))
```

```python
q1_index = len(data_Kaithi) // 4
q1 = data_Kaithi[q1_index][0]

q3_index = 3 * len(data_Kaithi) // 4
q3 = data_Kaithi[q3_index][0]

iqr_Kaithi = q3 - q1

print("Interquartile range:", iqr_Kaithi)
```
